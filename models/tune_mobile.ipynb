```python
from utils import read_data, evaluation, rollingCV, knn, random_forest
import pandas as pd
from IPython.display import display
```

```python
RD = read_data.ReadData()
Eval = evaluation.Evaluation()
hd = RD.read_Hd_Mobile()
K = 8
RCV = rollingCV.RollingCV(hd[::-1], K, 30, 14, 30)
KNN = knn.KNN(Eval, 60)
RandomForest = random_forest.RandomForest(Eval)
points_of_sales = sorted(hd.id.unique())
```

### KNN (all points)

```python
parameters = pd.DataFrame()
for k in range(1, K+1):
    train = RCV.getKthTrainSet(k, isML=True, isRegression=True)
    params = KNN.tune_parameters(train)
    params['k'] = k
    parameters = pd.concat([parameters, params], ignore_index=True)
display(parameters)
parameters.to_csv('parameters_mobile/knn_all.csv')
```

### Random Forest (all points)

```python
parameters = pd.DataFrame()
for k in range(1, K+1):
    train = RCV.getKthTrainSet(k, isML=True, isRegression=True)
    params = RandomForest.tune_parameters(train)
    params['k'] = k
    parameters = pd.concat([parameters, params], ignore_index=True)
display(parameters)
parameters.to_csv('parameters_mobile/rf_all.csv')
```
